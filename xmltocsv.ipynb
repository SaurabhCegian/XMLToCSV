```
import pandas as pd
import itertools
from bs4 import BeautifulSoup as b
```

```
with open("/content/physsim-network.xml", "r") as f:
    content = f.read()

soup = b(content, 'html.parser')
def nodepart(soup):
  mainTag = soup.find_all('node')
  Id = []
  x = []
  y = []
  for repElem in mainTag:
    Id.append(repElem.get('id'))
    x.append(repElem.get('x'))
    y.append(repElem.get('y'))
  data = [item for item in itertools.zip_longest(Id, x, y)] 
  return pd.DataFrame(data=data, columns= ['id', 'x', 'y'])

def linkpart(soup):
  mainTag = soup.find_all('link')
  Id = []
  From = []
  to = []
  length = []
  freespeed = []
  capacity = []
  permlanes = []
  oneway = []
  modes = []
  for repElem in mainTag:
      Id.append(repElem.get('id'))
      From.append(repElem.get('from'))
      to.append(repElem.get('to'))
      length.append(repElem.get('length'))
      freespeed.append(repElem.get('freespeed'))
      capacity.append(repElem.get('capacity'))
      permlanes.append(repElem.get('permlanes'))
      oneway.append(repElem.get('oneway'))
      modes.append(repElem.get('modes'))

  data = [item for item in itertools.zip_longest(Id,From,to,length,freespeed,capacity,permlanes,oneway,modes)] 
  return pd.DataFrame(data=data, columns= ['id', 'from', 'to','length', 'freespeed', 'capacity', 'permlanes', 'oneway', 'modes'])

left = nodepart(soup)
right = linkpart(soup)
```

```
df = pd.merge(left, right, how ='inner', on ='id') 
df.to_csv("csvfile.csv",index=False )
df.head(10)
```
